```python

def convert_to_dataframe(x):
    return pd.DataFrame(x)
```

```python
def create_dataset(dataset, look_back):
    dataX, dataY = [], []
    for i in range(len(dataset) - look_back - 1):
        # Create a sequence of `look_back` elements
        a = dataset[i:(i + look_back)]
        dataX.append(a)
        # The target value is the next element after the sequence
        dataY.append(dataset[i + look_back])
    return np.array(dataX), np.array(dataY)
```

```python
df = pd.read_csv(r'C:\Users\whiter04\OneDrive - Johnson Matthey\00. cspark\ELY\ELY-TimeSeries\Data\master_data_frame.csv')
df_copy = df.copy()
df_copy = df_copy.drop('Unnamed: 0',axis=1) 
df_copy


to_drop = ['temp_anode_inlet_set', 'temp_cathode_inlet_set', 'pressure_anode_set',
           'pressure_cathode_set', 'flow_anode_di_water_supply_set', 'flow_cathode_di_water_supply_set', 
          'flow_cathode_di_water_supply', 'signal_in_o2_in_h2_sensor']

df_copy = df_copy.drop(to_drop, axis=1)
df_copy


df_copy['seconds'] = (df_copy['time_running_total']/1000).astype(int)
# df_copy['minutes'] = (df_copy['seconds']/60).astype(int)

# df_copy['hours'] = (df_copy['minutes']/60).astype(int)
# df_copy['days'] = (df_copy['hours']/24).astype(int)
# df_copy['weeks'] = (df_copy['days']/7).astype(int)
df_copy = df_copy.drop('time_running_total',axis=1)

# df_copy = df_copy.tail(150000)
df_copy
```

```python
import matplotlib.pyplot as plt

time_period = 'seconds'



df_time = df_copy.groupby(time_period).mean().reset_index()

df_time = df_time.tail(75000)

# df_time.set_index('seconds',inplace = True)

plt.plot(df_time['signal_in_h2_in_o2_sensor'])

plt.show()




downsampled_df = df_time.head(75000) 



downsampled_df = downsampled_df[['signal_in_h2_in_o2_sensor']]


downsampled_df = np.array(downsampled_df)
```

```python

pipeline = Pipeline([
    ('scaler', MinMaxScaler()),
    # Add more preprocessing steps as needed
])

# Fit and transform the data
transformed_data = pipeline.fit_transform(downsampled_df)


transformed_data



```

```python
seq_length = 5
X, y = create_dataset(transformed_data, seq_length)

train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

```

```python
X_train_reshaped = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], X_train.shape[2]))

X_test_reshaped = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], X_test.shape[2]))
```

```python
X_train_reshaped.shape
```

```python
model = Sequential()

model.add(LSTM(50,input_shape=(X_train_reshaped.shape[1], X_train_reshaped.shape[2]), return_sequences=True))
model.add(LSTM(50, return_sequences=False))
model.add(Dense(25, activation='relu'))  # Dense layer with ReLU activation
model.add(Dense(1))  # Output layer with no activation (linear activation)

model.compile(loss='mean_absolute_error', optimizer='adam')
model.fit(X_train_reshaped, y_train, epochs=3, batch_size=1, verbose=1)
```
